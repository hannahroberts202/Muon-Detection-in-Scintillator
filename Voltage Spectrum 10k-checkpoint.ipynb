```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

def read_files(directory):
    peak_voltages = []
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            df = pd.read_csv(os.path.join(directory, filename), usecols=[1], header=None, skiprows=21)
        elif filename.endswith(('.xlsx', '.xls')):
            df = pd.read_excel(os.path.join(directory, filename), usecols=[1], header=None, skiprows=21)
        else:
            continue

        peak_voltage = df[1].abs().max()
        if np.isfinite(peak_voltage):
            peak_voltages.append(peak_voltage)
        else:
            print(f"Non-finite peak voltage found in {filename}")
    return peak_voltages

def plot_voltage_histogram(peak_voltages):
    if not peak_voltages:
        print("No valid peak voltages to plot.")
        return
    print(f"Number of peak voltages: {len(peak_voltages)}")  # Debug: Print the number of voltages
    plt.figure(figsize=(10, 6))
    plt.hist(peak_voltages, bins=50, edgecolor='black')
    plt.title('Histogram of Peak Voltages')
    plt.xlabel('Peak Voltage (V)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

current_directory = os.getcwd()
peak_voltages = read_files(current_directory)
plot_voltage_histogram(peak_voltages)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import pylandau
from scipy.stats import gaussian_kde

# Function to read files and extract peak voltages
def read_files(directory):
    peak_voltages = []
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            df = pd.read_csv(os.path.join(directory, filename), usecols=[1], header=None, skiprows=21)
        elif filename.endswith(('.xlsx', '.xls')):
            df = pd.read_excel(os.path.join(directory, filename), usecols=[1], header=None, skiprows=21)
        else:
            continue

        peak_voltage = df[1].abs().max()
        if np.isfinite(peak_voltage):
            peak_voltages.append(peak_voltage)
        else:
            print(f"Non-finite peak voltage found in {filename}")
    return peak_voltages

# Function to plot histogram with Landau fitting
def plot_voltage_histogram_with_landau_fit(peak_voltages):
    if len(peak_voltages) == 0:
        print("No valid peak voltages to plot.")
        return

    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))

    # Estimate the density using gaussian_kde
    kde = gaussian_kde(peak_voltages)
    x = np.linspace(min(peak_voltages), max(peak_voltages), 1000)
    kde_values = kde(x)  # Define kde_values here

    # Creating the density plot
    sns.kdeplot(peak_voltages, bw_adjust=0.5, fill=True)

    # Adjusted estimates for MPV and eta based on the density plot
    # You may need to manually adjust these values based on your histogram
    mpv_adjusted = np.percentile(peak_voltages, 25)  # Try a lower percentile for a better initial estimate
    eta_adjusted = (np.percentile(peak_voltages, 75) - np.percentile(peak_voltages, 25)) / 8 # Smaller division factor to reduce the width


    # Scale the amplitude
    amplitude_adjusted = max(kde_values)*1.3 # The amplitude may need to be adjusted after MPV and eta

    # Use the adjusted parameters to plot the Landau distribution
    y_landau_adjusted = amplitude_adjusted * pylandau.landau(x, mpv=mpv_adjusted, eta=eta_adjusted)

    plt.plot(x, y_landau_adjusted, 'r-', lw=2, label='Adjusted Landau fit')

    # Setting the title and labels
    plt.title('Density and Landau Fit of Peak Voltages')
    plt.xlabel('Peak Voltage (V)')
    plt.ylabel('Density')
    plt.legend()

    plt.show()
# Using the function to read files and plot the histogram with Landau fitting
current_directory = os.getcwd()
peak_voltages = np.array(read_files(current_directory))
plot_voltage_histogram_with_landau_fit(peak_voltages)
```
